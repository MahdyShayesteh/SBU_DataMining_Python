#initial lib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
import tensorflow as tf
import cv2 as cv
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from google.colab import files
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image 
```

# load images and add one dimention to pic

```python
url="https://i.pinimg.com/originals/5f/34/d4/5f34d4cfce3233c12e572f8f2a3b56b6.jpg"
image = io.imread(url) 
image = np.expand_dims(image,0)
plt.imshow(image[0])
```

# define function to show augmented pictures

```python
def plotImages(images_arr):
  fig,axes=plt.subplots(2,5,figsize=(20,20))
  axes = axes.flatten()
  for img,ax in zip(images_arr,axes):
    ax.imshow(img)
    ax.axis('off')
  plt.tight_layout()
  plt.show()
```

# image augmented

```python
# define this variable to set all augmented type to picture 
gen = ImageDataGenerator(rotation_range=10,width_shift_range=0.1,height_shift_range=0.,shear_range=0.15,zoom_range=0.1,channel_shift_range=10,horizontal_flip=True)
```

```python
# generate batches of augmented image 
aug_iter = gen.flow(image)
```

```python
# get 10 sample of augmented images
aug_images=[next(aug_iter)[0].astype(np.uint8) for i in range(10)]
```

```python
# plot augmented images in 2 row and 5 column
plotImages(aug_images)
```
